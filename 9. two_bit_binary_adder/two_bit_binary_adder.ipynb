```python
#Build a binary two-bit adder using quantum logic gates
```

```python
from qiskit import *
```

```python
circuit = QuantumCircuit(4, 2)
circuit.x(0) # initalize input A =1
circuit.x(1) # initalize input B =1
circuit.barrier() #Do not reorder or optimize gates across this point.  # Prevent compiler reordering; separates logic stages
#It tells the transpiler:
#“Everything before this must stay before”
#“Everything after this must stay after”

circuit.cx(0, 2) # First cx (cnot) Use first input qubit  as its controlled to conditionally flip the sum qubit. If A = 1, flip SUM
circuit.cx(1, 2) # Second cx (cnot) we use second input qubit as its control for flipping the sum. If B = 1, flip SUM
#If only one of the inputs is one then the sum will be flipped once ending up in the one state.


#If both inputs are one then the both CNOT gate would be triggered, flipping the sum from zero o one and then from one back to zero.
circuit.ccx(0, 1, 3) # # Flip carry-out only if both A and B are 1. ccx gate to use first two input qubits as controlls and flip the fourth carryout qubit accordangly.
circuit.barrier()
circuit.measure(2, 0) # measure SUM
circuit.measure(3, 1) # measure CARRY-OUT
circuit.draw()
```
