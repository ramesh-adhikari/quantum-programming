```python
# The CCX gate (also called the Toffoli gate)
# For three qubits

# control qubits:  a,b

# target qubit: c
# Rule:
# If either control qubit is |0⟩, do nothing to the target.
# If both control qubits are |1⟩, flip the target qubit (apply an X gate).

# In other words
# Control qubit 1 = |0⟩ OR control qubit 2 = |0⟩  → target unchanged
# Control qubit 1 = |1⟩ AND control qubit 2 = |1⟩ → apply X to the target
```

```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector, plot_histogram, array_to_latex ,plot_state_qsphere
```

```python
circuit = QuantumCircuit(3) # initialize circuit with 3 qubits
circuit.draw()
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
circuit.remove_final_measurements()  # no measurements allowed
statevector = Statevector(circuit) # Create a Statevector object representing the quantum state of the circuit
array_to_latex(statevector.data)
plot_bloch_multivector(statevector) # Visualize the quantum state on the Bloch sphere for each qubit
```

```python
circuit = QuantumCircuit(3) # initialize circuit with 3 qubits
# TO apply CNOT Gate
circuit.ccx(0,1,2) # first parameter is control qubit index by 0 and second parameter is targit qubit index by 1
circuit.draw()
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
circuit.remove_final_measurements()  # no measurements allowed
statevector = Statevector(circuit) # Create a Statevector object representing the quantum state of the circuit
print(statevector.data)
plot_bloch_multivector(statevector) # Visualize the quantum state on the Bloch sphere for each qubit
```

```python
# Both qubits are in zero state, since the control qubit was zero, nothing happend and still in the initiall zero state
#Now lets insert X gate to flip that control to one just before cnot
```

```python
circuit = QuantumCircuit(3) # initialize circuit with 3 qubits
circuit.x(0) # flip the first qubit from 0 to 1
# TO apply CNOT Gate
circuit.ccx(0,1,2) # first parameter is control qubit index by 0 and second parameter is targit qubit index by 1
circuit.draw()
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
circuit.remove_final_measurements()  # no measurements allowed
statevector = Statevector(circuit) # Create a Statevector object representing the quantum state of the circuit
print(statevector.data)
plot_bloch_multivector(statevector) # Visualize the quantum state on the Bloch sphere for each qubit
```

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
circuit.remove_final_measurements()  # no measurements allowed
statevector = Statevector(circuit) # Create a Statevector object representing the quantum state of the circuit
print(statevector.data)
plot_bloch_multivector(statevector) # Visualize the quantum state on the Bloch sphere for each qubit
```
